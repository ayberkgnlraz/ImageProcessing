```python
import cv2

image = cv2.imread('image_thres1.jpg')

# Convert the image to grayscale for better edge detection
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply edge detection (using Canny)
edges = cv2.Canny(gray, 50, 150)

# Find contours in the edged image
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
print(contours)

# Assuming the largest contour is the rectangle
largest_contour = max(contours, key=cv2.contourArea)

# Get the bounding rectangle around the contour
x, y, w, h = cv2.boundingRect(largest_contour)

# Calculate dimensions in centimeters using DPI
cm_per_inch = 2.54
length_cm = (w / 300) * cm_per_inch
width_cm = (h / 300) * cm_per_inch

print(f"Length of rectangle: {length_cm:.2f} cm")
print(f"Width of rectangle: {width_cm:.2f} cm")
```

```python
import cv2

# Known conversion factor: pixels to centimeters
pixels_to_cm = 0.1  # Adjust this based on your actual conversion factor

# Open a connection to the camera (usually camera index 0 is the default built-in camera)
cap = cv2.VideoCapture(0)

# Check if the camera opened successfully
if not cap.isOpened():
    print("Error: Could not open camera.")
    exit()

# Create a window for the live stream
cv2.namedWindow('Live Stream')

while True:
    # Read a frame from the camera
    ret, frame = cap.read()

    # Check if the frame was successfully read
    if not ret:
        print("Error: Could not read frame.")
        break

    # Convert the frame to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Apply binary thresholding
    ret, thresh = cv2.threshold(gray, 100, 255, cv2.THRESH_BINARY)

    # Detect contours on the binary image using cv2.CHAIN_APPROX_NONE
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

    # Draw contours on the frame
    cv2.drawContours(frame, contours, -1, (0, 255, 0), 2)

    # Measure the length of each detected contour
    for contour in contours:
        length_pixels = cv2.arcLength(contour, True)
        length_cm = length_pixels * pixels_to_cm
        print(f"Length: {length_cm:.2f} cm")

    # Display the frame in the 'Live Stream' window
    cv2.imshow('Live Stream', frame)

    # Break the loop if 'Esc' key is pressed
    if cv2.waitKey(1) == 27:
        break

# Release the camera and close the window
cap.release()
cv2.destroyAllWindows()
```
